```python
import numpy as np
import pandas as pd
#import hvplot.pandas  # noqa

#For Monte Carlo
import random

#Visualization
#import holoviews as hv

from tqdm.notebook import tqdm

#Historical Data
from stockstats import StockDataFrame as sdf
import yfinance as yf
from matplotlib import pyplot as plot 
```

```python
msft = yf.Ticker("fcx")
hist = msft.history(period="1y")
```

```python
def ottieni_macd(df):
    hist = sdf.retype(df)
    hist[['macd','macds', 'macdh']]
    ss = hist.macdh.values
    ss_1 = hist.macdh.shift(1).values
    h = pd.DataFrame(ss).rename(columns={0:'macdh'}).join(pd.DataFrame(ss_1).rename(columns={0:'macdh_1'}))
    signal = []

    for index, row in h.iterrows():
        if (row.macdh > row.macdh_1) and (row.macdh * row.macdh_1 < 0):
            signal.append(1)
        elif (row.macdh < row.macdh_1) and (row.macdh * row.macdh_1 < 0):
            signal.append(-1)
        else:
            signal.append(0)
    
    h['signal'] = signal

    # create figure and axis objects with subplots()
    fig,ax = plot.subplots(figsize=(50,30) )
    # make a plot
    ax.plot(hist.index, hist[['close']], color="red")
    
    # twin object for two different y-axis on the sample plot
    ax2=ax.twinx()
    # make a plot with different y-axis using second axis object
    ax2.plot(hist.index, h.signal,color="blue")
    
    plot.show()
    # save the plot as a file
```

```python
ottieni_macd(hist)
```

```python
def ottieni_kama(df):
    hist = sdf.retype(df)
    hist._get_kama('close', 10, fasts=34, slows=5)
    hist._get_kama('close', 10, fasts=34, slows=2)
    hist[['close','close_10_kama_34_2', 'close_10_kama_34_5']].plot(figsize=(30,20))
```

```python
ottieni_kama(hist)
```

```python
#import numpy as np
#import pandas as pd
from stockstats import StockDataFrame as sdf
import yfinance as yf
from matplotlib import pyplot as plt 

def RSI_periodo(ticker):
    df = yf.download(tickers=str(ticker),period='14d',interval='1d')
    df = sdf.retype(df)
    df.get('rsi')
    # plot price
    plt.figure(figsize=(15,5))
    plt.plot(df.index, df['adj close'])
    plt.title('Price chart (Adj Close)')
    plt.show()
    
    
    # plot correspondingRSI values and significant levels
    plt.figure(figsize=(15,5))
    plt.title('RSI chart')
    plt.plot(df.index, df['rsi'])
    
    plt.axhline(0, linestyle='--', alpha=0.1)
    plt.axhline(20, linestyle='--', alpha=0.5)
    plt.axhline(30, linestyle='--')
    
    plt.axhline(70, linestyle='--')
    plt.axhline(80, linestyle='--', alpha=0.5)
    plt.axhline(100, linestyle='--', alpha=0.1)
    plt.show()
```

```python
RSI_periodo('FCX')
```
